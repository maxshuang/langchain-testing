```python
from PIL import Image
import base64
import io

# Load the image
image_path = '/Users/maxhuang/projects/llm/langchain-testing/langchain/multimodal-test.jpeg'
image = Image.open(image_path)

# Convert image to base64
buffered = io.BytesIO()
image.save(buffered, format="JPEG")
image_base64 = base64.b64encode(buffered.getvalue()).decode('utf-8')
```

```python
local_model = "/Users/maxhuang/projects/llm/models/Qwen/Qwen2.5-Coder-14B-Instruct-GGUF/qwen2.5-coder-14b-instruct-q4_0.gguf"
```

```python
from langchain_community.llms import LlamaCpp
from langchain_core.prompts import ChatPromptTemplate

# Initialize the model
llm = LlamaCpp(
    model_path=local_model,
    n_ctx=250000,
    temperature=0.5,
    n_gpu_layers=8,
    n_batch=300
)

# Create the prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "Describe the image provided."),
    ("user", [
        {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{image_base64}"}}
    ])
])

# Generate the description
response = llm(prompt)
print(response)
```
